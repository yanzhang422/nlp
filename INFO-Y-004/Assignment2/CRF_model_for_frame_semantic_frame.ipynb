```python
def _read_tsv(input_file, quotechar=None):
    """Reads a tab separated value file."""
    with open(input_file, "r") as f:
      reader = csv.reader(f, delimiter="\t", quotechar=quotechar)
      lines = []
      for line in reader:
        lines.append(line)
      return lines
```

get all data from the corpus

```python
#read all tsv
import os
domains = ['antiquite','archwiki','histclio','histwiki','prehistoire']
from tqdm import tqdm
```

```python
dt = []
frames = []
for domain in tqdm(domains):
    print('reading data from domain:{%s}'%(domain))
    list_files = os.listdir(os.path.join('calor_frames',domain)) #list all files
    for file in list_files:
        lines = _read_tsv(os.path.join('calor_frames',domain,file))
        for line in lines:
            if len(line)==0:continue
            if 'TARGET' in line[-1]:
                tars = line[-1].split(':')
                frames.append(tars[1])
                dt.append((line[3],line[4],line[5],line[9]))
```

first we need select three frames for CRFs, one frame with a high frequency, 
one with a medium frequency and 
one with a low frequency

according to Figure 2 in the background paper
select Attack, Finish_competition,Seeking

```python
select_frames = ['Attack','Finish_competition','Seeking']
```

```python
count_frames={'Attack':0,'Finish_competition':0,'Seeking':0}
for line in frames:
    for sl in select_frames:
        if sl==line:
            count_frames[sl]+=1
```

generate training and test data for each frame

for each frame lexicon unit,
we extract features including lemma,postag,dependency tag,

```python
def word2features(sent):
    word = sent[0]
    lemma=sent[1]
    postag = sent[2]
    dependency=sent[3]
    features = {
        'word.lower()': word.lower(),
        'word[-3:]': word[-3:],
        'word[-2:]': word[-2:],
        'word.isupper()': word.isupper(),
        'word.istitle()': word.istitle(),
        'word.isdigit()': word.isdigit(),
        'postag': postag,
        'postag[:2]': postag[:2],
        'depend':dependency,
        'depend[:2]':dependency[:2]
    }
    return features
```

Negative samples for downsampling 1:5 can continue to fine tune

```python
import numpy as np
_neg_idx = np.arange(len(_x_neg))
_neg_idx = np.random.choice(_neg_idx,len(_x_pos)*5)
```

```python
_x_neg = [_x_neg[idx] for idx in _neg_idx]
_x = [fea for line in _x_pos] + [fea for line in _x_neg]
_y = [['Attack'] for line in _x_pos] + [['Other'] for line in _x_neg]
```

```python
train_features = [[word2features(lex)] for lex in _x]
```

```python
len(train_features),len(_y)
```

```python
#train valid split
from sklearn.cross_validation import StratifiedKFold
```

```python
#we random chose train and valid data
```

The training set and verification set ratio is 8:2

```python
skf = StratifiedKFold(np.array(_y).reshape(-1),n_folds=5, shuffle=True, random_state=1024)
for ind_tr, ind_te in skf:
    break
```

```python
train_x = [train_features[idx] for idx in ind_tr]
test_x = [train_features[idx] for idx in ind_te]
train_y = [_y[idx] for idx in ind_tr]
test_y = [_y[idx] for idx in ind_te]
```

```python
crf = sklearn_crfsuite.CRF(
    algorithm='lbfgs',
    c1=0.1,
    c2=0.1,
    max_iterations=100,
    all_possible_transitions=True
)
crf.fit(train_x, train_y)
```

```python
labels = list(crf.classes_)
labels
```

```python
y_pred = crf.predict(test_x)
metrics.flat_f1_score(test_y, y_pred,
                      average='weighted', labels=labels)
```

```python
f1_score = metrics.flat_f1_score(test_y, y_pred,
                      average='weighted', labels=labels)
per_score = metrics.flat_precision_score(test_y, y_pred,
                      average='weighted', labels=labels)
recal_score = metrics.flat_recall_score(test_y, y_pred,
                      average='weighted', labels=labels)
```

```python
print('the attack model performance: %s %s %s'%(f1_score,per_score,recal_score))
```
